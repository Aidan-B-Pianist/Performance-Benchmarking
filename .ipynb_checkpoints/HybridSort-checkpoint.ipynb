```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
sns.set_context("notebook")

palette = sns.color_palette("flare")
sns.set_palette(palette)

k_values = [10, 20, 30, 40, 50]

# Average times in ms across array sizes (10,000, 25,000, 50,000, 75,000, 100,000)
avg_times_ms = [
    (5.16 + 15.853 + 43.6331 + 47.1644 + 61.4168) / 5,  # K=10
    (4.9586 + 19.2491 + 50.8242 + 48.6413 + 67.4422) / 5,  # K=20
    (5.7837 + 15.5628 + 34.959 + 51.5144 + 68.7506) / 5,  # K=30
    (6.3224 + 17.0465 + 38.2713 + 55.1804 + 75.4997) / 5,  # K=40
    (7.202 + 18.951 + 40.4069 + 58.9307 + 81.7057) / 5   # K=50
]


plt.figure(figsize=(10, 5))
plt.plot(k_values, avg_times_ms, marker='o', color='darkorange')
plt.title("Hybrid Sort - Average Runtime vs K, (Random Input)")
plt.xlabel("K-value")
plt.ylabel("Average Runtime (ms)", rotation = 0, labelpad = 75)
plt.grid(True)
plt.show()


#This graph shows the correlation between time and varying k-value with the legend varying n-value
#We see that the run time for k is best at 30, but when we ran tests on 30, the data shows that 20 was the more optimal k-value
```

```python
array_sizes = [10000, 25000, 50000, 75000, 100000]

hybrid_sort_times = [4.9586, 19.2491, 50.8242, 48.6413, 67.4422]  # k = 20
quick_sort_times = [5.3561, 15.1166, 32.1266, 48.34, 64.5983]
insertion_sort_times = [1206.3385, 6649.339, 26352.9087, 56796.3312, 102181.5785]

plt.figure(figsize=(10, 5))
plt.plot(array_sizes, hybrid_sort_times, marker='o', label='Hybrid Sort (k=20)')
plt.plot(array_sizes, quick_sort_times, marker='o', label='Quick Sort')
plt.plot(array_sizes, insertion_sort_times, marker='o', label='Insertion Sort')
plt.title("Sorting Algorithms - Runtime vs Array Size, (Random Input)")
plt.xlabel("Array Size")
plt.ylabel("Runtime (ms)", rotation = 0, labelpad = 45)
plt.legend()
plt.yscale('log')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()

#The optimal k value is 20 as the hybrid line intersects with the fourth orange dot (quick sort line) at
#hybrid sort making it slightly faster than normal quicksort
```

```python
#K-values
k_values = [10, 20, 30, 40, 50]

# Average times in ms across array sizes (100, 250, 500, 750, 1000)
avg_times_ms = [
    (0.2681 + 1.4162 + 5.8081 + 14.0272 + 22.8918) / 5, # K=10
    (0.2230 + 1.4041 + 6.3858 + 13.3131 + 23.8134) / 5, # K=20
    (0.2031 + 1.3733 + 5.9906 + 14.0108 + 23.7989) / 5, # K=30
    (0.2080 + 1.5186 + 6.0384 + 16.3927 + 23.3772) / 5, # K=40
    (0.1828 + 1.4053 + 6.1078 + 13.3913 + 22.4692) / 5, # K=50
]

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(k_values, avg_times_ms, marker='o', color='darkorange')
plt.title("Hybrid Sort - Average Runtime vs K, (Pre-sorted Input)")
plt.xlabel("K-value")
plt.ylabel("Average Runtime (ms)", rotation = 0, labelpad = 75)
plt.grid(True)
plt.show()

#This graph shows the best K-value for the pre-sorted input
#As we can see the K-value is best at 50 because we already have pre sorted input and insertion sort works better with pre sorted values.
```

```python
array_sizes = [100, 250, 500, 750, 1000]

#(All in ms)
hybrid_sort_times = [0.223, 1.4041, 6.3858, 13.3131, 23.8134]
quick_sort_times = [0.2313, 1.5219, 5.7783, 13.3549, 23.826]
insertion_sort_times = [0.001, 0.0026, 0.0042, 0.0059, 0.0079]

#Plot
plt.figure(figsize=(10, 5))
plt.plot(array_sizes, hybrid_sort_times, marker='o', label='Hybrid Sort (k=20)')
plt.plot(array_sizes, quick_sort_times, marker='o', label='Quick Sort')
plt.plot(array_sizes, insertion_sort_times, marker='o', label='Insertion Sort')
plt.title("Sorting Algorithms - Runtime vs Array Size, (Pre-sorted Input)")
plt.xlabel("Array Size")
plt.ylabel("Runtime (ms)", rotation = 0, labelpad = 45)
plt.legend()
plt.yscale('log')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()

#This graph shows the sorting algorithm runtime against array size using pre sorted arrays as input. As we can see, it's significantly slower than 
#if we were to randomly generate values.
```
